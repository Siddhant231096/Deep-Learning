# NLTK ----------------------------------

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk import tokenize
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.stem import PorterStemmer
```

```python
stopw=stopwords.words('english')
```

```python
data="I am great and I want to rule on  whole world sometime"
```

```python
# seperate on behave of space 

wdata=word_tokenize(data)   
```

```python
useful_data=[i for i in wdata if i.lower() not in stopw]
```

```python
useful_data
```

```python
# steamming 

ps=PorterStemmer()
for i in useful_data:
    print(ps.stem(i))
```

```python
wt=WordNetLemmatizer()
for i in useful_data:
    print(wt.lemmatize(i))
```

# Tweets analysis

```python
twwt=word_tokenize(str(tw))
twuse=[i for i in twwt if i.lower() not in stopw]
```

```python
for i in twuse:
    print("PorterStemmer -->> ",ps.stem(i))
    
```

```python
for i in twuse:
    print("WordNetLemmatizer -->> ",wt.lemmatize(i))
```

```python
for i in twuse:
    print(i," --- ",wt.lemmatize(i)," --- ",ps.stem(i))
```

# Now applying Sentimental analysis




```python
from textblob import TextBlob
import matplotlib.pyplot as plt
```

```python
positive=0
negative=0
nutral=0
for tweet in tweets:
    analysis=TextBlob(tweet.text)
    sent=analysis.sentiment
    if sent.polarity<0:
        negative=negative+1
    elif sent.polarity==0:
        nutral=nutral+1
    elif sent.polarity>0:
        positive=positive+1
        
print('Positive ',positive)
print('Negativity ',negative)
print('Nutral ',nutral)
```

```python
plt.xlabel('polarity')
plt.ylabel('count')
plt.bar(['positive','neagative','nural'],[positive,negative,nutral])
```
